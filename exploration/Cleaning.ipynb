```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt

import seaborn as sns
import numpy as np
%matplotlib notebook
```

```python
path = Path('../data')
```

```python
def import_one_file(path):
    df = pd.read_csv(path);
    df['Timestamp'] = pd.to_datetime(df['Timestamp'])
    df = df.set_index('Timestamp')
    return df

def import_data(path):
    return import_one_file(path/'market_data.csv'), import_one_file(path/'trades.csv')

market, trades = import_data(path)
```
